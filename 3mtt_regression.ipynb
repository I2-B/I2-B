```python
#importing the libaries needed

import pandas as pd
#splittng of model
from sklearn.model_selection import train_test_split
#import the algorthims
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression

#PREPROCESSING LIBRIES

from sklearn.preprocessing import StandardScaler
#metrics for regression
from sklearn.metrics import mean_squared_error
```

```python
df  = pd.read_csv("/content/petrol_consumption.csv")
df
```

```python
df.columns
```

```python
X = df.drop('Petrol_Consumption', axis =1)
y = df['Petrol_Consumption']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.3,random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#standardized the dataset

ramlat = StandardScaler()
X_train = ramlat.fit_transform(X_train)
X_test = ramlat.transform(X_test)
```

```python
#model buliding
L_model = LinearRegression() #linaer model
R_model = RandomForestRegressor() #random forest
D_model = DecisionTreeRegressor() #decesion tree model
K_model = KNeighborsRegressor()  #KNN model
```

```python
#fitting the models
L_model.fit(X_train,y_train)
l_pred = L_model.predict(X_test)

#random forest
R_model.fit(X_train,y_train)
r_pred = R_model.predict(X_test)

#decession tree
D_model.fit(X_train,y_train)
d_pred = D_model.predict(X_test)

#knn
K_model.fit(X_train,y_train)
k_pred = K_model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, l_pred) # linear model mean squared error
print("the mean squared error for the model is:", mse)
mse_r = mean_squared_error(y_test, r_pred) # random forset  model mean squared error
print("the mean squared error for the model is:", mse_r)
mse_d = mean_squared_error(y_test, d_pred)# decision tree model mean squared error
print("the mean squared error for the model is:", mse_d)
mse_k = mean_squared_error(y_test, k_pred)# KNN model mean squared error
print("the mean squared error for the model is:", mse_k)
```

```python

from sklearn.metrics import mean_absolute_error
mse = mean_absolute_error(y_test, l_pred) # linear model mean absolaute s error
print("the mean squared error for the model is:", mse)
```
